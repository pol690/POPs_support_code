# Data parser for OCP compounds measured with Agilent GCMS

#### The code in the notebook takes a .csv file of the "complete summary" of measurements, conducted on Musshunter software. It removes the need for a special .xml file to create a custom output summary.

```python
import pandas as pd
```

```python
OCP_redact= pd.read_csv("/media/artemk/Data/POPs/OCP data/AG2-20190111_Artem/AG2-20190111_Artem/AG2-2019011.csv")
```

```python
OCP_redact
```

```python
columns_to_drop = [0,1,3,4,5,6]
```

```python
OCP_redact.drop(OCP_redact.columns[columns_to_drop], axis=1, inplace=True)
```

```python
OCP_redact.set_index("Unnamed: 2")
```

```python
columns_names = OCP_redact.columns.get_values().tolist()
```

```python
new_columns_names = [name for name in columns_names if "Unnamed" not in name]
```

```python
new_columns_names = [s.strip(' Results') for s in new_columns_names]
```

```python
len(new_columns_names)
```

```python
newer_columns_names = ["Sample name"] + new_columns_names
```

```python
OCP_redact = OCP_redact[OCP_redact.columns[::2]]
```

```python
OCP_redact.columns = newer_columns_names
OCP_redact.drop([0])
```

```python
OCP_redact.to_excel("/media/artemk/Data/POPs/OCP data/AG2-20190111_Artem/AG2-20190111_Artem/AG2-2019011.xlsx")
```
